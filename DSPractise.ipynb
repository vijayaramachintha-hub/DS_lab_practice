```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the probability of success for the Bernoulli distribution
p = 0.3

# Generate 1000 Bernoulli trials
data = np.random.binomial(1, p, 1000)

# Plot histogram
plt.hist(data, bins=2, edgecolor='black')

# Add labels and title
plt.title("Histogram of Bernoulli Distribution (p = 0.3)")
plt.xlabel("Outcome")
plt.ylabel("Frequency")

# Set x-axis ticks
plt.xticks([0, 1])

# Display the graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameters for the binomial distribution
n = 10
p = 0.5

# Generate 1000 Binomial trials
data = np.random.binomial(n, p, 1000)

# Plot histogram
sns.histplot(data, kde=True, stat="density", bins=n+1)

# Add labels and title
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xlabel("Number of Successes")
plt.ylabel("Density")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameters for the binomial distribution
lam = 4
# Generate 1000 Binomial trials
data = np.random.poisson(lam, 1000)

# Plot histogram
sns.histplot(data, kde=False, stat="density", bins=range(min(data),max(data)+1))

# Add labels and title
plt.title(f"Poisson Distribution (lamda={lam})")
plt.xlabel("Number of Events")
plt.ylabel("Density")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameters for the binomial distribution
mu = 0
sigma = 1
# Generate 1000 Binomial trials
data = np.random.normal(mu,sigma,1000)

# Plot histogram
sns.histplot(data, kde=True, stat="density")

# Add labels and title
plt.title(f"Normal Distribution (mean={mu},std={sigma})")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
